```python
from sklearn import linear_model
import matplotlib.pyplot as matplt     
import pandas as pd                 
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import mean_squared_error


def encode(x):
    return 1 if x == 'Partial' else 0

def show_ggplot(train):
    matplt.style.use(style='ggplot')
    matplt.rcParams['figure.figsize'] = (10, 6)
    matplt.hist(train.SalePrice, color='magenta')
    matplt.show()
    
def show_column(train):
    target_col = np.log(train.SalePrice)
    matplt.hist(target_col, color='maroon')
    matplt.show()

def show_saleprice_garagearea(train_val) :
    matplt.scatter(x=train_val['GarageArea'], y=train_val.SalePrice)
    matplt.ylabel('Sale Price')
    matplt.xlabel('Garage Area')
    matplt.show()
    
def show_saleprice_garagearea_without_outliers(train_val) :
    matplt.scatter(x=train_val['GarageArea'], y=np.log(train_val.SalePrice))
    matplt.xlim(-200, 1600)  # This forces the same scale as before
    matplt.ylabel('Sale Price')
    matplt.xlabel('Garage Area')
    matplt.show()

def get_null_count(train_val):
    column_count = pd.DataFrame(train_val.isnull().sum().sort_values(ascending=False)[:25])
    column_count.columns = ['Null Count']
    column_count.index.name = 'Feature'
    print(column_count)

def show_category(train_val):
    categoricals = train_val.select_dtypes(exclude=[np.number])
    print(categoricals.describe())
    
def show_sale_condition_median_sales_price(train_val):
    condition_pivot = train_df.pivot_table(index='SaleCondition', values='SalePrice', aggfunc=np.median)
    condition_pivot.plot(kind='bar', color='blue')
    matplt.xlabel('Sale Condition')
    matplt.ylabel('Median Sale Price')
    matplt.xticks(rotation=0)
    matplt.show()

def show_en_condition_median_sales_price(train_val):
    condition_pivot = train_val.pivot_table(index='enc_condition', values='SalePrice', aggfunc=np.median)
    condition_pivot.plot(kind='bar', color='gold')
    matplt.xlabel('Encoded Sale Condition')
    matplt.ylabel('Median Sale Price')
    matplt.xticks(rotation=0)
    matplt.show()
    
def get_train_model(x_train_val, y_train_val):
    lr = linear_model.LinearRegression()
    train_model = lr.fit(x_train_val, y_train_val)
    return train_model

def show_linear_regression_model(train_predictions, actual_values):
        matplt.scatter(train_predictions, actual_values, alpha=.75,color='black') 
        matplt.xlabel('Predicted Price')
        matplt.ylabel('Actual Price')
        matplt.title('Linear Regression')
        matplt.show()

def show_ridge_regularization_alpha(alpha_val, x_train_val, y_train_val, x_test_val, y_test_val):
            rm = linear_model.Ridge(alpha=alpha_val)
            ridge_model = rm.fit(x_train_val, y_train_val)
            preds_ridge = ridge_model.predict(x_test_val)

            matplt.scatter(preds_ridge, actual_values, alpha=.75, color='maroon')
            matplt.xlabel('Predicted Price')
            matplt.ylabel('Actual Price')
            matplt.title('Ridge Regularization with alpha = {}'.format(alpha_val))
            overlay = 'R^2 is: {}\nRMSE is: {}'.format(
                ridge_model.score(x_test_val, y_test_val),
                mean_squared_error(y_test_val, preds_ridge))
            matplt.annotate(s=overlay, xy=(12.1, 10.6), size='x-large')
            matplt.show()

if __name__ == "__main__":
    test_df = pd.read_csv('test.csv')
    train_df = pd.read_csv('train.csv')

    print("Train data shape:", train_df.shape)
    print("Test data shape:", test_df.shape)

    show_ggplot(train_df)
    
    show_column(train_df)
    
    numericFeatures = train_df.select_dtypes(include=[np.number])
    
    corr = numericFeatures.corr()

    show_saleprice_garagearea(train_df)

    # create a new dataframe with some outliers removed
    train_df = train_df[train_df['GarageArea'] < 1200]
    show_saleprice_garagearea_without_outliers(train_df)

    get_null_count(train_df)

    show_category(train_df)

    train_df['enc_street'] = pd.get_dummies(train_df.Street, drop_first=True)
    test_df['enc_street'] = pd.get_dummies(test_df.Street, drop_first=True)

    show_sale_condition_median_sales_price(train_df)

    train_df['enc_condition'] = train_df.SaleCondition.apply(encode)
    test_df['enc_condition'] = test_df.SaleCondition.apply(encode)
    
    show_en_condition_median_sales_price(train_df)
    
    data = train_df.select_dtypes(include=[np.number]).interpolate().dropna()

    sales_log = np.log(train_df.SalePrice)
    actual_data = data.drop(['SalePrice', 'Id'], axis=1)
    
    x_train_val, x_test_val, y_train_val, y_test_val = train_test_split(actual_data, sales_log, random_state=42, test_size=.33)

    train_model = get_train_model(x_train_val, y_train_val)

    train_predictions = train_model.predict(x_test_val)

    actual_values = y_test_val
    show_linear_regression_model(train_predictions, actual_values)
    

    for i in range(-2, 3):
        alpha_val = 10 ** i
        show_ridge_regularization_alpha(alpha_val, x_train_val, y_train_val, x_test_val, y_test_val)

    #Creating final Data
    #Creating Data
    final_df = pd.DataFrame()
    final_df['Id'] = test_df.Id

    
    test_feats = test_df.select_dtypes(include=[np.number]).drop(['Id'], axis=1).interpolate()

    # Train Prediction
    train_predictions = train_model.predict(test_feats)
    print("Train Predictions are: \n", train_predictions[:10], "\n")

    #Exponential Prediction
    final_train_predictions = np.exp(train_predictions)
    print("Final Train Predictions are: \n", final_train_predictions[:10])

    # assign these train_predictions and check
    final_df['SalePrice'] = final_train_predictions
    print(final_df.head())

    final_df.to_csv('finalreport.csv', index=False)
```
